```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/refs/heads/main/Dados/imigrantes_canada.csv'
df = pd.read_csv(url)
df
```

```python
df.info()
```

```python
df.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
```

```python
brasil
```

```python
brasil_dict = {'ano':brasil.index.tolist(), 'imigrantes': brasil.values.tolist() }
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
def grafico_brasil():
  plt.figsize = (8,4)
  plt.title('Imigração do Brasil para o Canadá')
  plt.xlabel('Ano')
  plt.ylabel('Número de imigrantes')
  plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
  plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
  plt.show()

grafico_brasil()
```

```python
argentina = df.loc['Argentina', anos]
```

```python
argentina_dict = {'ano' : argentina.index.tolist(), 'imigrantes': argentina.values.tolist()}
argentina_dict
```

```python
df_argentina = pd.DataFrame(argentina_dict)
df_argentina
```

```python
def grafico_argentina():
  plt.plot(df_argentina['ano'], df_argentina['imigrantes'])
  plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
  plt.title('Imigrantes Argentina para o Canadá')
  plt.xlabel('Ano')
  plt.ylabel('Imigrantes')
  plt.show()
```

```python
grafico_argentina()
grafico_brasil()
```

```python
plt.plot(df_argentina['ano'], df_argentina['imigrantes'], label= 'Argentina')
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'], label= 'Brasil')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'],)
plt.title('Comparação Brasil e Argentina')
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
plt.legend()
plt.show()
```

```python
fig, axis = plt.subplots(figsize=(8,4))
axis.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3, color='g')
axis.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', fontsize=18, loc='left')
axis.set_xlabel('Ano', fontsize=14)
axis.set_ylabel('Iimigrantes', fontsize=14)
axis.xaxis.set_tick_params(labelsize=12)
axis.yaxis.set_tick_params(labelsize=12)
axis.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15,5))
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Imigrantes')
axs[1].grid()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigração dos 4 maiores paises da América do Sul para o Canada\n1980 a 2013')

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
  ax.set_xlabel('Ano')
  ax.set_ylabel('Imigrantes')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

```python
america_sul = df.query('Região =="América do Sul"')
```

```python
america_sul
```

```python
america_sul_sorted = america_sul.sort_values('Total', ascending=True)


cores = []
for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title("América do Sul: Brasil foi o quarto país com mais Imigrantes\npara o Canadá entre 1980 e 2013", loc='left', fontsize=18)
ax.set_ylabel('')
ax.set_xlabel('Imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

fig.savefig('imigracao_america_sul.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()

```
